```python
import numpy as np
import seaborn as sns
from bokeh.io import output_file, show, curdoc
from bokeh.plotting import figure
```

```python
# load dataset car_crashes

car_df = sns.load_dataset("car_crashes")
car_df
```

```python
# create numpy array from dataset

car_arr = np.array(car_df)
car_arr
```

```python
# create new numpy array with only 3 columns

car_arr3 = car_arr[:,:3]
car_arr3
```

```python
class Zscore:

    def __init__(self, np_columns: np.ndarray):
        self.car_arr3 = car_arr3

    def get_score(self, x):
        arr = []
        for i in x:
            z = abs((i - np.mean(x)) / np.std(x))
            arr.append(z)

        return np.array(arr)
```

```python
scores = Zscore(car_arr3)
```

```python
def get_average_score(x, y, z):
    return np.average([x,y,z])
```

```python
# calculate z-score for each of column

z_total = scores.get_score(car_arr3[:,0])
z_total
```

```python
z_speed = scores.get_score(car_arr3[:, 1])
z_speed
```

```python
z_alcohol = scores.get_score(car_arr3[:, 2])
z_alcohol
```

```python
# calculate average z-score for every car crash and create new array to save this data
vfunc = np.vectorize(get_average_score, otypes=[float])
out = vfunc(z_total, z_alcohol, z_speed)
out
```

```python
# choose top z-score
top = 7
# create new np array where crashes with top z-score are set to False and thr rest to True
k = np.where(out >= np.sort(out)[-top:][0], False, out)
k
```

```python
c = np.where(k != False, True, k)
c
```

```python
# create new array with our data and column that identify outliers
new_arr = np.column_stack((car_arr3, c))
new_arr
```

```python
# visualize
while True:
    reason = ['total', 'speeding', 'alcohol']
    ch = int(input('If you want visualize data choose 1, to exit type smth else'))
    if ch == 1:
        y = int(input('Please input y axis \n'
              'If '+ reason[0] + ' input 0 \n'
              'If '+ reason[1] + ' input 1 \n'
              'If '+ reason[2] + ' input 2 \n'))
        x = int(input('Please input x axis \n'
              'If '+ reason[0] + ' input 0 \n'
              'If '+ reason[1] + ' input 1 \n'
              'If '+ reason[2] + ' input 2 \n'))

        plot = figure(title = "Statistic of car crashes", match_aspect = True)
        plot.title.align = "center"
        plot.title.text_font = "times"
        plot.title.text_font_style = "italic"
        plot.title.text_font_size = "25px"
        plot.xaxis.axis_label = reason[x]
        plot.yaxis.axis_label = reason[y]
        plot.axis.minor_tick_in = -3
        plot.axis.minor_tick_out = 6


        plot.scatter(new_arr[:, x][np.where(new_arr[:, -1] == False)], new_arr[:, y][np.where(new_arr[:, -1] == False)], size = 9, color = '#800000', alpha = 0.8)
        plot.scatter(new_arr[:, x][np.where(new_arr[:, -1] == True)], new_arr[:, y][np.where(new_arr[:, -1] == True)], size = 9, color = '#428af5', alpha = 0.8)

        output_file('numpy_scatter1.html')
        curdoc().theme = 'dark_minimal'
        show(plot)
    else:
        break
```
